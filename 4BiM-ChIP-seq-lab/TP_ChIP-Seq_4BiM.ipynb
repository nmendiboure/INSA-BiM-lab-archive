### Fonction de lissage et réduction à moyenne glissante

```python
def lissage_reduction(x,y,p):
    '''Lissage et réduction par moyenne glissante sur 2p+1 points'''
    xout = np.empty(0) ; yout = np.empty(0)
    for index in range(p, len(y)-p+1, p):
        xout = np.append(xout,np.mean(x[index-p : index+p+1]))
        yout = np.append(yout,np.mean(y[index-p : index+p+1]))
    return xout,yout
```

 ### Fonction de chargement d'une séquence fasta

```python
def load_fasta_seq(filename):
    ''' Chargement sequence genomique a partir d'un fichier fasta'''
    seq = str()
    my_file = open(filename, "r") # ouverture fichier fasta
    for i in my_file.readlines(): # parcourt chaque ligne du fichier
        line=i.strip() # supprrime les retours à la ligne
        if line != '': #si la ligne n'est pas vide
            if line[0]!=">": #et si la ligne n'est pas un entete (les entêtes débutent par >)
                seq+=line
    my_file.close()
    return seq
```

#### Fonction de calcul du %GC

```python
def GC_content(seq, window=5000):
    ''' Calcul du taux de GC le long du genome '''
    x = [] ; GC = [] # x coordonnees des fenetres
    for i in range(0,len(seq), window//2):
        s = seq[i: i+window]  
        g = s.count('G') + s.count('g') # nb de G
        c = s.count('C') + s.count('c') # nb de C
        GC.append(float(g+c)/len(s))
        x.append(i+1) # coordonnee debut fenetre
    return x,GC
```

Il est primordial de tra-vailler avec des ratio d’enrichissement !

## Peak Calling 

Nous allons pouvoir passer à l’étape de "peak calling" (ou déconvolution) i.e. l’identiﬁcation des régions du génome ﬁxées par Fis. Toute la diﬃculté réside dans le fait de pouvoir distinguer le signal non spéciﬁque (évalué à l’aide du contrôle) du signal de ﬁxation protéique. Pour cela, nous allons calculer des scores d’enrichissement. En
général, des logiciels (dont les fondements algorithmiques et statistiques varient) sont utilisés pour cette étape et permettent la détection automatique des régions enrichies.

Nous allons réaliser cette étape nous-mêmes, qui va consister à :

+ (1) évaluer la variabilité des signaux 
+ (2) normaliser les signaux 
+ (3) calculer un rapport d’enrichissement de ﬁxation Fis par rapport au contrôle à chaque position
+ (4) estimer la ﬁabilité (intervalle de conﬁance)
+ (5) sélectionner et ﬁltrer les signaux les plus forts 

### 1) IC 95\% de la couverture à chaque posistion génomique, pour les deux signaux :

```python
# Signaux bruts et écart-type
#colonnes : 0 position, 1 couverture fis, 2 écart type fis, 3 couverture mock, 4 écart type mock 
signals = np.column_stack((pos,cov,np.sqrt(cov),covmock,np.sqrt(covmock)))
                           
# Calcul IC 95% (avant normalisation)
#colonnes : 0 position, 1 borne inférieure IC95% fis, 2 borne sup fis, 3 borne inf mock, 4 borne sup mock
IC = np.column_stack((pos,cov-2*signals[:,2],cov+2*signals[:,2],covmock-2*signals[:,4],covmock+2*signals[:,4]))
```

### 2) Normalisation des signaux

#### Normalisation des couvertures par la couverture totale :

```python
# ajout de 2 colonnes à la table signals : 5 couverture fis normalisée par la couverture total fis, 6 idem pour mock                          
signals = np.column_stack((signals,signals[:,1]/np.sum(cov), signals[:,3]/np.sum(covmock)))
```

```python
print(signals)
```

#### Calcul de l'enrichissement à chaque position

```python
# ajout de la colonne n°7 contenant le rapport à fis/mock à chaque position
signals = np.column_stack((signals,signals[:,5]/signals[:,6]))
```

```python
print(signals)
```

#### Estimation d’un IC 95% sur le rapport

```python
# Variance rapport
ratio_var = (signals[:,5]**2/signals[:,6]**2)*(((signals[:,1]/np.sum(cov)**2)/signals[:,5]**2) + ((signals[:,3]/np.sum(covmock)**2)/signals[:,6]**2))

# Ecart type rapport pour IC95
# ajout de la colonne n°8 contenant l'écart type sur le rapport
signals = np.column_stack((signals,np.sqrt(ratio_var)))

# Calcul de l IC95%
ICrapport = np.column_stack((pos,signals[:,7]-2*signals[:,8],signals[:,7]+2*signals[:,8]))
```

### 3) Calculer un rapport d’enrichissement de ﬁxation Fis par rapport au contrôle à chaque position

#### Elimination des positions pour lesquelles le signal est trop bruité 

```python
signals_filtered = np.copy(signals)

#On ignore les rapports bruités (on les met à 0) qui présentent soit : 

## moins de 9 reads dans le test ou dans le contrôle (signal pas significativement différent de 0)
signals_filtered[np.where((signals_filtered[:,1] < 9) | (signals_filtered[:,3] < 9)),7] = 0 

## un enrichissement trop faible (< 3) ou un IC trop grand (std > 0.5)
signals_filtered[np.where((signals_filtered[:,7] <= 3) | (signals_filtered[:,8] >= 0.5)),7] = 0
```

#### Détermination des centres des pics de fixation

```python
window=200 
pos_strong_sites = [] # position du pic
ratio_strong_sites = [] #rapport d'enrichissement fis/mock associé au pic
# on parcourt le génome en fenêtres de 200 pb pour sélectionner les maximum locaux
for i in range(0, int(signals_filtered[-1,0]-window),window):  #signals_filtered[-1,0] retourne la dernière position
    local_signal = signals_filtered[i:i+window,] # signaux de la fenêtre de 200pb
    idxmax = np.argmax(local_signal[:,7]) # indice du rapport fis/mock le plus grand de la fenêtre
    if local_signal[idxmax, 7] != 0: # si ce rapport n'est pas égal à 0, on le garde avec sa position
        pos_strong_sites.append(local_signal[idxmax,0])
        ratio_strong_sites.append(local_signal[idxmax,7])
        
print ("Nombre de sites forts identifiés en TP :", len(pos_strong_sites))
```

### Découpage du génome en fragments et comptage du nombre de signaux par fragment
+ Fonction de comptage

```python
def count_signals(sites,*args, **kwargs):
    '''
    Division du génome en n bins non chevauchantes et comptage du nombre de signaux a l'intérieur de chaque bin
    sites = coordonnees des sites de structure [[début, fin], ..., [début,fin]]
    '''
    nbins = kwargs.get("nbins",20) # nombre de bins
    lgenome = kwargs.get("lgenome", 4639675) # longueur du génome
    lbin = lgenome//nbins # longueur d'une bin
    lastbin = lgenome%nbins # longueur de la fraction restante

    bins = [] # liste des bins de structure [[début,fin,nb signaux], ..., [début,fin,nb signaux]]
    for i in range(1,lgenome-lbin,lbin): # initialisation des bins : on indique le début et la fin de chaque bin
        bins.append([i,i+lbin-1,0])
    
    if lastbin != 0: # ajout de la fraction restante 
        bins.append([bins[-1][1],bins[-1][1]+lastbin,0])

    bins = np.array(bins)
    for site in sites: # attribue chaque signal à un bin (la position du signal doit être dans l'intervalle du bin)
        bins[np.where((bins[:,0] <= site) & (bins[:,1] >= site)),2] += 1
    return bins
```
